```python
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 -Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\nColumn: {col}")
    print(f"\nNo of Outliers:{outliers.shape[0]}")
    print(f'Outlier values:\n{outliers[col].values}')
```

```python
#Explain the concept of standardization (z-score normalization).

# Ans:- Standardization, also known as Z-score normalization, is a data preprocessing technique used to scale numerical features so that they have a mean of 0 and a standard deviation of 1. This ensures that each feature contributes equally to the model's performance and prevents features with larger scales from dominating the learning process.
# It is particularly useful in algorithms that are sensitive to the scale of data, such as K-Nearest Neighbors, Support Vector Machines, and logistic regression. Standardization is performed by subtracting the mean of the feature and dividing by its standard deviation.
# This technique assumes that the data is approximately normally distributed and is widely used in scenarios where we want to maintain the effects of outliers without compressing the feature range.

```

```python
#Standardization of Numerical Variables
#We have to convert Categorical data into numerical data, then we can apply Standardization of our Numerical Variables

cat_col=df.select_dtypes(include=['category','object']).columns
df=pd.get_dummies(df,columns=cat_col,dtype=int)
df
```

```python
#Discuss the need for converting categorical data into dummy variables (one-hot encoding).

#Ans:- When working with machine learning models, especially those that rely on mathematical computations, categorical data must be converted into a numerical format. One common approach for this is one-hot encoding, which converts each category into a dummy variable â€” a binary column that indicates the presence (1) or absence (0) of a category.

# This is important because:

# Most ML algorithms cannot interpret text or category labels directly.

# Assigning numerical labels (e.g., Red = 1, Blue = 2) may introduce false ordinal relationships that don't exist.

# One-hot encoding ensures that each category is treated independently, avoiding misleading patterns in the data.

# For example, a "Color" feature with values Red, Green, and Blue becomes three columns: Color_Red, Color_Green, and Color_Blue, each containing 0 or 1.

# Thus, converting categorical data into dummy variables ensures that models can interpret and learn effectively from the data without introducing bias or incorrect assumptions.

```
